**1. Buat sebuah Array yang berisi data tinggi badan dan berat badan beberapa orang**

````python
import array as array
tinggi_badan = array.array('i', [164, 155, 161, 154, 158, 157, 178, 151, 168, 179])
berat_badan = array.array('i', [69, 66, 56, 65, 52, 79, 60, 52, 79, 63])
````

Hasil Display dengan Pandas

````python
import pandas as pd
tbtable = list(tinggi_badan)
tb = pd.DataFrame(tbtable, columns=["Tinggi Badan"])
tb
````

````python
import pandas as pd
bbtable = list(berat_badan)
bb = pd.DataFrame(bbtable, columns=["Berat Badan"])
bb
````

**2. Kemudian hitung rata-rata tinggi badan dan berat badan!**

````python
average_tinggi_badan = sum(tinggi_badan) / len(tinggi_badan)
average_berat_badan = sum(berat_badan) / len(berat_badan)
````

Hasil Rata-Rata

````python
print("Rata-rata tinggi badan:", average_tinggi_badan)
print("Rata-rata berat badan:", average_berat_badan)
````
