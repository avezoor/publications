# Implementasi Selection Sort

````python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
````

Menentukan Array

````python
arr = [64, 25, 12, 22, 11]
````

Panggil Program

````python
selection_sort(arr)
print(arr)
````

## Menghitung Kompleksitas Selection Sort

````python
def selection_sort(arr):                                # c₁ * 1
    n = len(arr)                                        # c₂ * 1
    for i in range(n - 1):                              # c₃ * (n-1) -> Loop variant: 0 ≤ i < n-1
        min_idx = i                                     # c₄ * (n-1)
        for j in range(i + 1, n):                       # c₅ * (∑ᵢ₌₁ⁿ⁻¹ (n - i - 1)) -> Loop variant: i+1 ≤ j < n
            if arr[j] < arr[min_idx]:                   # c₆ * (∑ᵢ₌₁ⁿ⁻¹ (n - i - 1))
                min_idx = j                             # c₇ * (∑ᵢ₌₁ⁿ⁻¹ (n - i - 1))
        arr[i], arr[min_idx] = arr[min_idx], arr[i]     # c₈ * (n-1)
````

Notasi

$
T_n = c_1 \times 1 + c_2 \times 1 + c_3 \times (n-1) + c_4 \times (n-1)
+ c_5 \times \sum_{i=1}^{n-1} (n-i-1)
+ c_6 \times \sum_{i=1}^{n-1} (n-i-1)
+ c_7 \times \sum_{i=1}^{n-1} (n-i-1)
+ c_8 \times (n-1)
$

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ (c_5 + c_6 + c_7) \sum_{i=1}^{n-1} (n-i-1)
$

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ (c_5 + c_6 + c_7) \sum_{i=0}^{n-2} i
$

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ (c_5 + c_6 + c_7) \times \frac{(n-1)(n-2)}{2}
$

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ \frac{(c_5 + c_6 + c_7)}{2} (n-1)(n-2)
$



Identifikasi derajat tertinggi
- Suku konstanta  
$(c_1 + c_2)$  adalah  $\mathcal{O}(1).$  

- Suku linear  
$(c_3 + c_4 + c_8)(n-1)$  adalah  $\mathcal{O}(n).$

- Suku kuadratik  
$\frac{(c_5 + c_6 + c_7)}{2} (n-1)(n-2)$  adalah  $\mathcal{O}(n^2).$  

Jadi didapatkan

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ \frac{(c_5 + c_6 + c_7)}{2} (n-1)(n-2)
$

$
T_n = \mathcal{O}(n^2)
$

## **Analisis Best Case dan Worst Case**

## Best Case (Data Sudah Terurut)
Contoh Input :

```
A = [1, 2, 3, 4, 5]
```

1. i = 0*, `min_idx = 0`, memeriksa `[2, 3, 4, 5]`, `min_idx` tetap 0, tidak ada pertukaran.
2. i = 1, `min_idx = 1`, memeriksa `[3, 4, 5]`, `min_idx` tetap 1, tidak ada pertukaran.
3. i = 2, `min_idx = 2`, memeriksa `[4, 5]`, `min_idx` tetap 2, tidak ada pertukaran.
4. i = 3, `min_idx = 3`, hanya `[5]` yang tersisa, tidak ada pertukaran.

### Perhitungan
- Jumlah Perbandingan:

    $$
    (n-1) + (n-2) + ... + 1 = \frac{(n-1)n}{2} = O(n^2)
    $$

- Jumlah Pertukaran: `0` (karena tidak ada elemen yang harus dipindahkan)
- Kompleksitas Waktu: $O(n²)$

## Worst Case (Data Terurut Terbalik)
Contoh Input


```
A = [5, 4, 3, 2, 1]
```
1. i = 0, `min_idx = 0`, menemukan elemen terkecil `1` di indeks 4, tukar `A[0] ↔ A[4]` → `[1, 4, 3, 2, 5]`.
2. i = 1, `min_idx = 1`, menemukan elemen terkecil `2` di indeks 3, tukar `A[1] ↔ A[3]` → `[1, 2, 3, 4, 5]`.
3. i = 2, `min_idx = 2`, tidak ada pertukaran (karena `3` sudah di tempatnya).
4. i = 3, `min_idx = 3`, tidak ada pertukaran (karena `4` sudah di tempatnya).

### Perhitungan
-  Jumlah Perbandingan:  
    $$
    O(n^2)
    $$
- Jumlah Pertukaran: `(n-1)` kali (karena setiap elemen harus ditukar)
- Kompleksitas Waktu: $O(n²)$



## Ringkasan Kompleksitas Waktu

| Kasus        | Contoh Input       | Perbandingan | Pertukaran  | Kompleksitas Waktu |
|-------------|--------------------|-------------|------------|--------------------|
| Best Case  | `[1, 2, 3, 4, 5]` | $ O(n^2) $ | $ 0 $ | $ O(n^2) $ |
| Worst Case | `[5, 4, 3, 2, 1]` | $ O(n^2) $ | $ O(n) $ | $ O(n^2) $ |


### Kesimpulan
- Selection Sort membutuhkan $ O(n^2) $ dalam perbandingan, baik dalam base case maupun worst case.
- Perbedaannya hanya pada jumlah swap, yang minimal di best case (0 pertukaran) dan maksimal di worst case ($ O(n^2) $ pertukaran).
- Cocok untuk array pendek atau kasus di mana jumlah swap sedikit.
