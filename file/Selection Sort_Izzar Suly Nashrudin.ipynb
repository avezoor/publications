# Selection Sorting Implementation

````python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
````

Specifying an Array

````python
arr = [64, 25, 12, 22, 11]
````

Call Program

````python
selection_sort(arr)
print(arr)
````

## Calculating Selection Sort Complexity

````python
def selection_sort(arr):                                # c₁ * 1
    n = len(arr)                                        # c₂ * 1
    for i in range(n - 1):                              # c₃ * (n-1) -> Loop variant: 0 ≤ i < n-1
        min_idx = i                                     # c₄ * (n-1)
        for j in range(i + 1, n):                       # c₅ * (∑ᵢ₌₁ⁿ⁻¹ (n - i - 1)) -> Loop variant: i+1 ≤ j < n
            if arr[j] < arr[min_idx]:                   # c₆ * (∑ᵢ₌₁ⁿ⁻¹ (n - i - 1))
                min_idx = j                             # c₇ * (∑ᵢ₌₁ⁿ⁻¹ (n - i - 1))
        arr[i], arr[min_idx] = arr[min_idx], arr[i]     # c₈ * (n-1)
````

Notation

$
T_n = c_1 \times 1 + c_2 \times 1 + c_3 \times (n-1) + c_4 \times (n-1)
+ c_5 \times \sum_{i=1}^{n-1} (n-i-1)
+ c_6 \times \sum_{i=1}^{n-1} (n-i-1)
+ c_7 \times \sum_{i=1}^{n-1} (n-i-1)
+ c_8 \times (n-1)
$

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ (c_5 + c_6 + c_7) \sum_{i=1}^{n-1} (n-i-1)
$

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ (c_5 + c_6 + c_7) \sum_{i=0}^{n-2} i
$

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ (c_5 + c_6 + c_7) \times \frac{(n-1)(n-2)}{2}
$

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ \frac{(c_5 + c_6 + c_7)}{2} (n-1)(n-2)
$



Identify the highest degree
- Constant terms  
$(c_1 + c_2)$ is $\mathcal{O}(1).$  

- Linear term  
$(c_3 + c_4 + c_8)(n-1)$ is $\mathcal{O}(n).$

- Quadratic term  
The quadratic term $\frac{(c_5 + c_6 + c_7)}{2}(n-1)(n-2)$ is $\mathcal{O}(n^2).$  

So we get

$
T_n = (c_1 + c_2) + (c_3 + c_4 + c_8)(n-1)
+ \frac{(c_5 + c_6 + c_7)}{2}(n-1)(n-2)
$

$
T_n = \mathcal{O}(n^2)
$

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
n = np.arange(1, 101, 1) # n from 1 to 100
T_n = (n * (n - 1)) / 2 # Execution time T(n) = (n(n-1))/2

# Big-O, Big-Theta, and Big-Omega
O_n2 = n**2 # O(n^2) - Upper bound
Omega_n2 = (1/4) * n**2 # Ω(n^2) - Lower bound
Theta_n2 = (1/2) * n**2 # Θ(n^2) - Tight bound

# Plot
plt.figure(figsize=(8, 6))
plt.plot(n, O_n2, label=r"$\mathcal{O}(n^2)$ (Upper bound)", linestyle="-", color="red")
plt.plot(n, Omega_n2, label=r"$\Omega(n^2)$ (Lower Limit)", linestyle="-", color="blue")
plt.plot(n, Theta_n2, label=r"$\Theta(n^2)$ (Tight Limit)", linestyle="-", color="purple")
plt.plot(n, T_n, label=r"$T(n) = \frac{(n-1)n}{2}$", linestyle="-", color="green")

# Labels and Titles with LaTeX
plt.xlabel(r"Number of Data $(n)$", fontsize=12)
plt.ylabel(r"Execution Time", fontsize=12)
plt.title(r"Plot of $\mathcal{O}(n^2)$, $\Omega(n^2)$, $\Theta(n^2)$, and $T(n)$ for Selection Sort", fontsize=14)

# Legend with LaTeX
plt.legend(fontsize=11)
plt.grid(True)

# Show Plot
plt.show()
````

## **Best Case and Worst Case Analysis**

## Best Case (Sorted Data)
Input Example:

```
A = [1, 2, 3, 4, 5]
```

1. i = 0*, `min_idx = 0`, checks `[2, 3, 4, 5]`, `min_idx` remains 0, no exchange.
2. i = 1, `min_idx = 1`, check `[3, 4, 5]`, `min_idx` remains 1, no exchange.
3. i = 2, `min_idx = 2`, check `[4, 5]`, `min_idx` remains 2, no exchange.
4. i = 3, `min_idx = 3`, only `[5]` remains, no exchange.

### Calculation
- Number of Comparisons:

    $$
    (n-1) + (n-2) + ... + 1 = \frac{(n-1)n}{2} = O(n^2)
    $$

- Number of Exchanges: `0` (since there are no elements to move)
- Time Complexity: $O(n²)$

## Worst Case (Reverse Sorted Data)
Input Example


```
A = [5, 4, 3, 2, 1]
```
1. i = 0, `min_idx = 0`, finds the smallest element `1` at index 4, swap `A[0] ↔ A[4]` → `[1, 4, 3, 2, 5]`.
2. i = 1, `min_idx = 1`, finds the smallest element `2` at index 3, swap `A[1] ↔ A[3]` → `[1, 2, 3, 4, 5]`.
3. i = 2, `min_idx = 2`, no swap (since `3` is already in place).
4. i = 3, `min_idx = 3`, no exchange (because `4` is already in place).

### Calculation
- Number of Comparisons:  
    $$
    O(n^2)
    $$
- Number of Exchanges: `(n-1)` times (since every element must be swapped)
- Time Complexity: $O(n²)$


## Time Complexity Summary

| Case | Input Example | Comparison | Exchange | Time Complexity |
|-------------|--------------------|-------------|------------|--------------------|
| Best Case | `[1, 2, 3, 4, 5]` | $ O(n^2) $ | $0 $ | $ O(n^2) $ |
| Worst Case | `[5, 4, 3, 2, 1]` | $ O(n^2) $ | $ O(n) $ | $ O(n^2) $ |


### Conclusion
- Selection Sort requires $O(n^2)$ in comparison, both in base case and worst case.
- The difference is only in the number of swaps, which is minimal in the best case (0 swaps) and maximal in the worst case ($O(n^2)$ swaps).
- Suitable for short arrays or cases where the number of swaps is small.
