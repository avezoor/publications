# **Visualisasi Data dengan Matplotlib**

```python
import matplotlib.pyplot as plt
```

## **Menyiapkan data yang akan di plot**


```python
import numpy as np
x = np.linspace(0,10,100)
y = x**2
```

## **Plot Data**

```python
plt.plot(x,y)
plt.show()
```

```python
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

## **Menambahkan warna dan marker**

```python
# Contoh plot grafik fungsi sin dan cos dengan color dan marker yang berbeda
x = np.arange(0,2,0.05)*np.pi
y = np.sin(x)
plt.plot(x,y)
y2 = np.cos(x)
plt.plot(x,y2,'r')
y3 = np.sin(2*x)
plt.plot(x,y3,'go-')
y4 = np.cos(2*x)
plt.plot(x,y4,marker='x',color='orange')
```

## **Menambahkan Legend**

```python
x = np.arange(0,2,0.05)*np.pi
y = np.sin(x)
plt.plot(x,y, label=r'$ \mathcal{y = sin(x)} $')
y2 = np.cos(x)
plt.plot(x,y2,'r', label=r'$ \mathcal{y = cos(x)} $')
plt.legend(loc='upper center')
plt.show()
```

## **Figure, Axes, Subplot**

```python
plt.figure()
ax1 = plt.axes()
ax1.plot(x,y)
ax1.set_title('Grafik sin(x)')
# Perintah di bawah akan menampilkan grafik pada figure yang berbeda
plt.figure()
plt.plot(x,2*y,'r')
plt.title('Grafik 2sin(x)')
plt.show()
```

```python
fig = plt.figure()
ax1 = plt.subplot(121) # plot 1 baris 2 kolom
ax1.plot(x,y)
ax1.set_title('Grafik sin(x)')
ax1.axis([0,6,-2,2])
ax1.set_xlabel('radian')
ax1.set_ylabel('sin(x)')
# Perintah di bawah ini akan menambahkan grafik di sebelah kanan grafik pertama
ax2 = plt.subplot(122)
ax2.plot(x,2*y,'r')
ax2.set_title('Grafik 2sin(x)')
ax2.axis([0,6,-2,2])
ax2.set_xlabel('radian')
ax2.set_ylabel('2sin(x)')
plt.show()
```

## **Latihan 1**

1. Cari bagaimana cara menampilkan Judul (title) dari grafik
2. Cari bagaimana cara memberikan label bada sumbu x dan sumbu y
3. Cari bagaimana cara menambahkan teks dalam grafik


```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat data
x = np.linspace(0, 10, 100)
y = np.exp(x)

# Membuat grafik
plt.plot(x, y)

# Menambahkan judul
plt.title("Grafik Exponensial dari x")

# Menambahkan label pada sumbu X dan Y
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")

# Menambahkan teks pada grafik
plt.text(2, 10, "Increasing", fontsize=12)

# Menampilkan grafik
plt.show()

# menyimpan Foto Plot
plt.savefig('plot.png')
```
